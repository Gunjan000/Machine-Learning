# Jiointplot grid
Smooth kernel density with marginal histograms

```python

#sns.set_theme(style="white")

#df = sns.load_dataset("penguins")

g = sns.JointGrid(data=data, x="total_bill", y="tip", space=0)
g.plot_joint(sns.kdeplot,
             fill=True, clip=((1, 51), (1, 10)),
             thresh=0, levels=100, cmap="rocket")
g.plot_marginals(sns.histplot, color="#03051A", alpha=1, bins=25)
```

# Distplot
fusion of Hiistogram, KDE/PDF, Rugplot

```python
tit= sns.load_dataset('titanic')
```

```python
tit.head(2)
```

```python
tit['age'].fillna(tit['age'].median(), inplace=True)
```

```python
sns.distplot(tit['age'])
```

```python
#ploting histogram
sns.distplot(tit['age'], bins=8, kde=False)
```

```python
#KDE
sns.distplot(tit['age'], hist=False)
```

```python
#RUGplot
sns.distplot(tit['age'], hist=False, kde=False, rug=True)
```

```python
sns.distplot(tit[tit['survived']==1]['age'], hist=False)
sns.distplot(tit[tit['survived']==0]['age'], hist=False)
```
